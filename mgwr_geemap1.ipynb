Membuka halaman Google Earth Engine dengan DAS Citarum sebagai titik tengahnya.

```python
#titik lokasi tengah DAS Citarum
Map = geemap.Map(center=(-6.3893832360229545,107.31268018020506), zoom=7)
Map
```

Membuka file output dari proses AGRMS dan menjadikan aset lokal di GEE Map

```python
namafile_hasil = os.path.join(folder_output, 'shp', 'output1.shp')
citarum_kecamatan = geemap.shp_to_ee(namafile_hasil)
```

Membuat tampilan peta sesuai warnanya.

```python
palette = ['fc0a0a', 'fc3232', 'f5a2a2', 'f7eded','d7d8fa','9396fa','4f54f7','4f54f7']
Map.add_styled_vector(citarum_kecamatan, column="mgwr_inter", palette=palette, layer_name="MGWR Intercept")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b1", palette=palette, layer_name="MGWR B1")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b2", palette=palette, layer_name="MGWR B2")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b3", palette=palette, layer_name="MGWR B3")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b4", palette=palette, layer_name="MGWR B4")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b5", palette=palette, layer_name="MGWR B5")
Map.add_styled_vector(citarum_kecamatan, column="mgwr_b6", palette=palette, layer_name="MGWR B6")

```
