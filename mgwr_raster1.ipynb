```python
import os
import sys
import time
import argparse
import geopandas as gpd
import fiona
import rasterio as rio
from rasterstats import zonal_stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mgwr.gwr import MGWR
from mgwr.sel_bw import Sel_BW
from mgwr.utils import shift_colormap, truncate_colormap
from datetime import datetime
```

```python
folder_root = os.getcwd()
print(folder_root);
```

```python
desa_polygon_file = os.path.join(folder_root, 'input', 'vektor', 'batas_desa_citarum_48S.shp')
print(desa_polygon_file)
gdf = gpd.read_file(desa_polygon_file)

crs = gdf.crs
gdf['x_utm'] = gdf.centroid.x
gdf['y_utm'] = gdf.centroid.y
```

```python
dem_raster_file = os.path.join(folder_root, 'input', 'raster', '2020_dem.tif')
raster_dem = rio.open(dem_raster_file, "r")  
profile = raster_dem.profile
transform = profile['transform']
nodata = raster_dem.nodata

img_dem = raster_dem.read(1)
img_dem = img_dem.astype('float32') 
img_dem[img_dem==nodata] = np.nan
zstats = zonal_stats(gdf, img_dem, affine=transform, prefix='dem_',nodata=nodata, stats='mean')
print(zstats)
vars = []
zstats_merged = []
zstats_merged.append(zstats) # zstats_merged is now a 2D list of lists [bands, shapes]
vars.append('dem_mean')
print(zstats_merged)
```
